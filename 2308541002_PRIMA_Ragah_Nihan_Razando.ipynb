```python
import math

print("MENCARI BILANGAN PRIMA DENGAN RENTANG 1000n SAMPAI 1000(n+1) JIKA NILAI n ADALAH DUA DIGIT TERAKHIR NIM KITA")
print("================================================================================================================================================")
# Mengatur nilai n sesuai NIM
n_02 = 2
a_02 = 1000 * n_02
b_02 = 1000 * (n_02 + 1)
print("Dua digit NIM akhir =", n_02)
print("Nilai 1000n =", a_02)
print("Nilai 1000(n+1) =", b_02)
print()

# Mencari bilangan prima pada rentang nilai a_02 hingga b_02
angka_prima_02 = []

for c_02 in range(a_02, b_02 + 1):
    if c_02 > 1:
        prima_02 = True
        for i_02 in range(2, int(math.sqrt(c_02)) + 1):
            if c_02 % i_02 == 0:
                prima_02 = False
                break
        if prima_02:
            angka_prima_02.append(c_02)

# Menampilkan hasil
print("Bilangan prima antara ",a_02," sampai dengan ",b_02," adalah")
print(angka_prima_02)
print("================================================================================================================================================")
print()
NIM_02 = int(input("Masukkan dua digit NIM akhir kalian "))
# Mengatur nilai n sesuai NIM
print("Dua digit NIM akhir anda =",NIM_02)
A_02 = 1000 * NIM_02
B_02 = 1000 * (NIM_02 + 1)
print("Nilai 1000n =", A_02)
print("Nilai 1000(n+1) =", B_02)
print()

# Mencari bilangan prima pada rentang nilai A_02 hingga B_02
Angka_Prima_02 = []

for C_02 in range(A_02, B_02 + 1):
    if C_02 > 1:
        Prima_02 = True
        for I_02 in range(2, int(math.sqrt(C_02)) + 1):
            if C_02 % I_02 == 0:
                Prima_02 = False
                break
        if Prima_02:
            Angka_Prima_02.append(C_02)

# Menampilkan hasil
print("Bilangan prima antara ",A_02," sampai dengan ",B_02," adalah")
print(Angka_Prima_02)
```
