```python
#Matriks dengan anggota 1, dengan baris dan kolom sesuai dengan yang diinput.

np.ones((2,2))#Matriks dengan anggota 1, dengan baris 2 dan kolom 2.
```

```python
#Matriks dengan anggota 1, dengan baris 10 dan kolom 10.

np.ones((10,10))
```

```python
#Matriks dengan anggota 0, dengan baris dan kolom sesuai dengan yang diinput.

np.zeros((4,3))#Matriks dengan anggota 0, dengan baris 4 dan kolom 3.
```

```python
#Matriks dengan anggota 0, dengan banyak matriks, baris dan kolom sesuai dengan yang diinput.

np.zeros((2,3,4))#Matriks dengan anggota 0, dengan banyak matriks sebanyak 2, baris 3 dan kolom 4.
```

```python
#Mencetak matriks identitas, dengan baris dan kolom sesuai dengan yang diinput.

np.identity(4)#Matriks identitas dengan ordo 4 x 4.
```

```python
#Mencetak matriks dengan baris, kolom, dan anggota sesuai dengan yang diinput.

np.full((4,5),7)#Matriks dengan baris 4, kolom 5, dan anggotanya 7.
```

```python
#Matriks dengan baris dan kolom sesuai dengan yang diinput, namun dengan anggota yang random.

np.random.rand(2,3)#Matriks dengan baris 2 dan kolom 3 dengan anggota random.
```

```python
#Matriks dengan anggota integer random dengan skala dan ukuran yang diinput

np.random.randint(1,10,size=(3,4))#Matriks dengan anggota integer 1-10 dengan ordo 3 x 4.
```

```python
rgh_02 = np.array([11,22,33,44])
print("Diketahui matriks rgh_02 sebagai berikut: rgh_02 =" , rgh_02)
print('_______________________________________________________________')

print("Penjumlahan matriks rgh_02 dengan 2")
print(rgh_02+2)
print('_______________________________________________________________')

print("Pengurangan matriks rgh_02 dengan 2")
print(rgh_02-2)
print('_______________________________________________________________')

print("Perkalian matriks rgh_02 dengan 2")
print(rgh_02*2)
print('_______________________________________________________________')

print("Pembagian matriks rgh_02 dengan 2")
print(rgh_02/2)
print('_______________________________________________________________')

print("Matriks rgh_02 dipangkatkan dengan 2")
print(rgh_02**2)
print('_______________________________________________________________')
```

```python
rgh_02 = np.array([11,22,33,44])
rnr_02 = np.array([1,2,3,4])
hasil_perkalian_matriks_02 = rgh_02*rnr_02
print("Diberikan matriks rgh_02 dan rnr_02 sebagai berikut:")
print("Matriks rgh_02=" , rgh_02)
print("Matriks rnr_02=" , rnr_02)
print('_______________________________________________________________')

print("Perkalian matriks rgh_02 dengan matriks rnr_02")
print(hasil_perkalian_matriks_02)
print('_______________________________________________________________')
```

```python
npwp_02 = np.array([[1,3,5,7],
                    [2,4,6,8],
                    [1,2,3,4]])
print(npwp_02)
```

```python
pwnp_02 = np.min(npwp_02)
print(pwnp_02)
```

```python
ppnw_02 = np.min(npwp_02,axis=0)
print(ppnw_02)
```

```python
wwnp_02 = np.min(npwp_02,axis=1)
print(wwnp_02)
```

```python
pppp_02 = np.max(npwp_02)
print(pppp_02)
```

```python
pppw_02 = np.max(npwp_02,axis=0)
print(pppw_02)
```

```python
wppp_02 = np.max(npwp_02,axis=1)
print(wppp_02)
```
