```python
baterai = float(input("Berapa banyak persentase baterai handphone anda? : " ))
if baterai == 100:
    print("Baterai handphone anda sudah penuh")
elif baterai < 100 and baterai >= 75:
    print("Handphone anda dapat digunakan dengan optimal")
elif baterai < 75 and baterai >= 50:
    print("Disarankan mengaktifkan 'Mode hemat daya'")
elif baterai < 50 and baterai >= 25:
    print("Handphone anda sudah tidak dapat digunakan secara optimal, harap hubungkan dengan pengisi daya")
elif baterai < 25 and baterai > 0:
    print("Hubungkan dengan pengisi daya")
```

```python
nama = input("Masukkan nama anda: ")
print("Halo Kak " + nama + ", Kami dari UNIQLO sedang mengadakan promo Ramadhan. Semakin banyak belanjaan anda, semakin besar diskonnya")
total_belanja = float(input("Total belanja: Rp. "))
diskon = 0
if total_belanja >= 500000:
    diskon = 0.3
elif total_belanja >= 250000:
    diskon = 0.2
elif total_belanja >= 100000:
    diskon = 0.1
elif total_belanja > 0:
    diskon = 0.05
jumlah_diskon = total_belanja * diskon
Total = total_belanja - jumlah_diskon
print("Sub Total: Rp.", total_belanja)
print("Diskon:", int(diskon * 100), "%")
print("Jumlah diskon: Rp.", jumlah_diskon)
print("Total Bill: Rp.", Total)
```
