```python
import numpy as np_02

def log_negatif_02(x_02):
    return np_02.log(abs(-x_02))

jalan_02 = True

while jalan_02:
    print('Diketahui x memenuhi: ')
    print('(x^2) + x - 1 = 0')
    print('')
    print('Tentukan nilai n yang memenuhi')
    print('(x^2024)-(x^2025)=(x*(1-x))^n')
    print('')
    print('penyelesaian:')
    print('Diketahui: ')
    print('a = 1')
    print('b = 1')
    print('c = -1')
    print('')
    print('Pertama-tama kami akan menghitung akar-akar penyelesaian persamaan kuadrat')
    print('substitusi ke dalam rumus (-b ± √(b^2-(4*a*c))) / 2*a')
    x_02 = str(input("Masukkan nilai penyelesaian yang ingin ditampilkan (positif/negatif/berhenti): "))
    print('')

    if x_02 == 'positif':
        # Nilai x positif
        x1_02 = (-1 + (np_02.sqrt(5))) / 2
        print('Nilai x1 adalah: {}'.format(x1_02))
        
        # Persamaan Kiri positif
        kiri_positif_02 = 2024 * np_02.log(x1_02) + np_02.log(1 - x1_02)
        
        # Persamaan Kanan positif
        kanan_positif_02 = x1_02 * (1 - x1_02)
        
        # Nilai n positif
        n_positif_02 = kiri_positif_02 / np_02.log(np_02.abs(kanan_positif_02))
        print(f"Nilai n yang memenuhi: {n_positif_02}")
        print('')
        
    elif x_02 == 'negatif':
        # Nilai x negatif
        x2_02 = (-1 - (np_02.sqrt(5))) / 2
        print('Nilai x2 adalah: {}'.format(x2_02))
        
        # Persamaan Kiri negatif
        kiri_negatif_02 = 2024 * log_negatif_02(x2_02) + np_02.log(1 - x2_02)
        
        # Persamaan Kanan negatif
        kanan_negatif_02 = x2_02 * (1 - x2_02)
        
        # Nilai n negatif
        n_negatif_02 = kiri_negatif_02 / log_negatif_02(np_02.abs(kanan_negatif_02))
        print(f"Nilai n yang memenuhi: {n_negatif_02}")
        print('')
        
    elif x_02 == 'berhenti':
        print('Program dihentikan. Terima kasih!')
        jalan_02 = False
        
    else:
        print('Silahkan memasukkan jawaban antara "positif", "negatif", atau "berhenti"')
        print('')
```
