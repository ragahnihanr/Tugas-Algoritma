```python
import cmath

# Menampilkan bilangan kompleks
z = 1 + 2j
print("Bilangan kompleks:", z)

# Menghitung akar kuadrat dari bilangan kompleks
sqrt_z = cmath.sqrt(z)
print("Akar kuadrat dari", z, "adalah", sqrt_z)

# Menghitung eksponensial dari bilangan kompleks
exp_z = cmath.exp(z)
print("Eksponensial dari", z, "adalah", exp_z)

# Menghitung logaritma natural dari bilangan kompleks
log_z = cmath.log(z)
print("Logaritma natural dari", z, "adalah", log_z)

# Menghitung sinus dari bilangan kompleks
sin_z = cmath.sin(z)
print("Sinus dari", z, "adalah", sin_z)

# Menghitung kosinus dari bilangan kompleks
cos_z = cmath.cos(z)
print("Kosinus dari", z, "adalah", cos_z)
```
