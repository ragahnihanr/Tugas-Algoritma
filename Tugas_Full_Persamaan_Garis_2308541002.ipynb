```python
import numpy as np
import matplotlib.pyplot as plt

#Input nilai x dan y pada garis pertama dan garis kedua
a_02 = int(input("Masukkan nilai x dari titik pertama pada garis pertama = "))
b_02 = int(input("Masukkan nilai y dari titik pertama pada garis pertama = "))
c_02 = int(input("Masukkan nilai x dari titik kedua pada garis pertama = "))
d_02 = int(input("Masukkan nilai y dari titik kedua pada garis pertama = "))

e_02 = int(input("Masukkan nilai x dari titik pertama pada garis kedua = "))
f_02 = int(input("Masukkan nilai y dari titik pertama pada garis kedua = "))
g_02 = int(input("Masukkan nilai x dari titik kedua pada garis kedua = "))
h_02 = int(input("Masukkan nilai y dari titik kedua pada garis kedua = "))

#Gradien garis pertama dan garis kedua
m1_02 = (d_02 - b_02) / (c_02 - a_02)
m2_02 = (h_02 - f_02) / (g_02 - e_02)

c1_02 = -m1_02 * a_02 + b_02  # Titik potong garis pertama terhadap sumbu y
c2_02 = -m2_02 * e_02 + f_02  # Titik potong garis kedua terhadap sumbu y 

# Persamaan garis pertama adalah y=m1_02(x-a_02)+b_02=m1_02x-m1_02a_02+b_02
# Persamaan garis kedua adalah y=m2_02(x-e_02)+f_02=m2_02x-m2_02e_02+f_02
print("Persamaan garis pertama adalah y={}x+{}".format(m1_02, c1_02))
print("Persamaan garis kedua adalah y={}x+{}".format(m2_02, c2_02))

if m1_02 == m2_02:
    print("Kedua garis sejajar sehingga tidak terdapat solusi atau terdapat tak berhingga banyaknya solusi")
    if c1_02 == c2_02:
        print("Kedua garis berhimpitan sehingga terdapat tak berhingga banyaknya solusi")
        x_02 = np.linspace(-5, 5, 150)
        plt.plot(x_02, m1_02 * (x_02 - a_02) + b_02, color='yellow')
        plt.plot(x_02, m2_02 * (x_02 - e_02) + f_02, color='green')
        plt.show()
    else:
        print("Kedua garis tidak berpotongan, sehingga tidak terdapat solusi.")
        x_02 = np.linspace(-5, 5, 150)
        plt.plot(x_02, m1_02 * (x_02 - a_02) + b_02, color='yellow')
        plt.plot(x_02, m2_02 * (x_02 - e_02) + f_02, color='green')
        plt.show()
else:
    print("Terdapat satu titik potong")
    tp_x_02 = (-m2_02 * e_02 + f_02 + m1_02 * a_02 - b_02) / (m1_02 - m2_02)
    tp_y_02 = m1_02 * (tp_x_02 - a_02) + b_02
    print("Titik potong kedua garis adalah ({},{})".format(tp_x_02, tp_y_02))
    
    x_02 = np.linspace(tp_x_02 - 5, tp_x_02 + 5, 150)
    plt.plot(x_02, m1_02 * (x_02 - a_02) + b_02, color='yellow')
    plt.plot(x_02, m2_02 * (x_02 - e_02) + f_02, color='green')
    plt.axvline(x=0, color='red', linestyle='--')
    plt.axhline(y=0, color='red', linestyle='--')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Garis melewati 2 titik
# 2. Garis diketahui gradien dan titik yang dilewati
# 3. Mengikuti persamaan garis ax+by=c dan dx+ey=f

# Akan dicari titik potong dari dua garis yg memenuhi salah satu dari ketiga kasus di atas

kasus_02 = int(input("Kasus berapakah yang akan dikerjakan? "))

if kasus_02 == 1:
    print("Garis melewati dua titik yang diketahui.")
    
    #Input nilai x dan y pada garis pertama dan garis kedua
    a_02 = int(input("Masukkan nilai x dari titik pertama pada garis pertama = "))
    b_02 = int(input("Masukkan nilai y dari titik pertama pada garis pertama = "))
    c_02 = int(input("Masukkan nilai x dari titik kedua pada garis pertama = "))
    d_02 = int(input("Masukkan nilai y dari titik kedua pada garis pertama = "))

    e_02 = int(input("Masukkan nilai x dari titik pertama pada garis kedua = "))
    f_02 = int(input("Masukkan nilai y dari titik pertama pada garis kedua = "))
    g_02 = int(input("Masukkan nilai x dari titik kedua pada garis kedua = "))
    h_02 = int(input("Masukkan nilai y dari titik kedua pada garis kedua = "))

    #Gradien garis pertama dan garis kedua
    m1_02 = (d_02 - b_02) / (c_02 - a_02)
    m2_02 = (h_02 - f_02) / (g_02 - e_02)

    c1_02 = -m1_02 * a_02 + b_02  # Titik potong garis pertama terhadap sumbu y
    c2_02 = -m2_02 * e_02 + f_02  # Titik potong garis kedua terhadap sumbu y garis kedua

    # Persamaan garis pertama adalah y=m1_02(x-a_02)+b_02=m1_02x-m1_02a_02+b_02
    # Persamaan garis kedua adalah y=m2_02(x-e_02)+f_02=m2_02x-m2_02e_02+f_02
    print("Persamaan garis pertama adalah y={}x+{}".format(m1_02, c1_02))
    print("Persamaan garis kedua adalah y={}x+{}".format(m2_02, c2_02))

    if m1_02 == m2_02:
        print("Kedua garis sejajar")
        if c1_02 == c2_02:
            print("Kedua garis berhimpitan sehingga terdapat tak berhingga banyaknya solusi")
            x_02 = np.linspace(-5, 5, 150)
            plt.plot(x_02, m1_02 * (x_02 - a_02) + b_02, color='yellow')
            plt.plot(x_02, m2_02 * (x_02 - e_02) + f_02, color='green')
            plt.axvline(x=0, color='red')
            plt.axhline(y=0, color='red')
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("Kedua garis tidak berpotongan, sehingga tidak terdapat solusi.")
            x_02 = np.linspace(-5, 5, 150)
            plt.plot(x_02, m1_02 * (x_02 - a_02) + b_02, color='yellow')
            plt.plot(x_02, m2_02 * (x_02 - e_02) + f_02, color='green')
            plt.axvline(x=0, color='red')
            plt.axhline(y=0, color='red')
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("Terdapat satu titik potong")
        tp_x_02 = (-m2_02 * e_02 + f_02 + m1_02 * a_02 - b_02) / (m1_02 - m2_02)
        tp_y_02 = m1_02 * (tp_x_02 - a_02) + b_02
        print("Titik potong kedua garis adalah ({},{})".format(tp_x_02, tp_y_02))
    
        x_02 = np.linspace(tp_x_02 - 5, tp_x_02 + 5, 150)
        plt.plot(x_02, m1_02 * (x_02 - a_02) + b_02, color='yellow', label="y={}(x-{})+{}".format(m1_02, a_02, b_02))
        plt.plot(x_02, m2_02 * (x_02 - e_02) + f_02, color='green', label="y={}(x-{})+{}".format(m2_02, e_02, f_02))
        plt.scatter([tp_x_02], [tp_y_02], color='blue', zorder=5)
        plt.text(tp_x_02, tp_y_02, '({}, {})'.format(tp_x_02, tp_y_02), fontsize=12)
        plt.axvline(x=0, color='red')
        plt.axhline(y=0, color='red')
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
        
elif kasus_02 == 2:
    print("Masing-masing garis diketahui gradien dan titik yang dilewati.")
    m1_02 = int(input("Masukkan gradien dari garis pertama = "))
    a_02 = int(input("Masukkan nilai x dari titik yang dilewati oleh garis pertama = "))
    b_02 = int(input("Masukkan nilai y dari titik yang dilewati oleh garis pertama = "))
    print("Persamaan garis pertama adalah y={}*(x-{})+{}".format(m1_02, a_02, b_02))
    
    m2_02 = int(input("Masukkan gradien dari garis pertama = "))
    c_02 = int(input("Masukkan nilai x dari titik yang dilewati oleh garis pertama = "))
    d_02 = int(input("Masukkan nilai y dari titik yang dilewati oleh garis pertama = "))
    print("Persamaan garis kedua adalah y={}*(x-{})+{}".format(m2_02, c_02, d_02))
    
    c1_02 = -m1_02 * a_02 + b_02
    c2_02 = -m2_02 * c_02 + d_02
    
    if m1_02 == m2_02:
        print("Kedua garis sejajar")
        if c1_02 == c2_02:
            print("Terdapat tak berhingga banyaknya solusi")
            x_02 = np.linspace(-5, 5, 150)
            plt.plot(x_02, m1_02 * (x_02 - a_02) + b_02, color='blue', label='Garis 1 & 2 (Berhimpitan)')
            plt.axvline(x=0, color='red', linestyle='--', linewidth=0.5)
            plt.axhline(y=0, color='red', linestyle='--', linewidth=0.5)
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("Tidak terdapat solusi")
            x_02 = np.linspace(-5, 5, 150)
            plt.plot(x_02, m1_02 * (x_02 - a_02) + b_02, color='yellow', label='Garis 1')
            plt.plot(x_02, m2_02 * (x_02 - c_02) + d_02, color='green', label='Garis 2')
            plt.axvline(x=0, color='red', linestyle='--', linewidth=0.5)
            plt.axhline(y=0, color='red', linestyle='--', linewidth=0.5)
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("Kedua garis berpotongan di satu titik.")
        tp_x_02 = (c2_02 - c1_02) / (m1_02 - m2_02)
        tp_y_02 = m1_02 * (tp_x_02 - a_02) + b_02
        print("Titik potong kedua garis adalah ({},{})".format(tp_x_02, tp_y_02))
        
        x_02 = np.linspace(tp_x_02 - 5, tp_x_02 + 5, 150)
        plt.plot(x_02, m1_02 * (x_02 - a_02) + b_02, color='yellow', label="Garis 1: y={}(x-{})+{}".format(m1_02, a_02, b_02))
        plt.plot(x_02, m2_02 * (x_02 - c_02) + d_02, color='green', label="Garis 2: y={}(x-{})+{}".format(m2_02, c_02, d_02))
        plt.scatter([tp_x_02], [tp_y_02], color='blue', zorder=5)
        plt.text(tp_x_02, tp_y_02, '({}, {})'.format(tp_x_02, tp_y_02), fontsize=12, ha='right')
        plt.axvline(x=0, color='red', linestyle='--', linewidth=0.5)
        plt.axhline(y=0, color='red', linestyle='--', linewidth=0.5)
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

elif kasus_02 == 3:
    print("Kedua garis memenuhi persamaan berbentuk ax+by=c dan dx+ey=f.")
    a_02 = int(input("Masukkan koefisien x pada garis pertama = "))
    b_02 = int(input("Masukkan koefisien y pada garis pertama = "))
    c_02 = int(input("Masukkan konstanta garis pertama = "))
    
    d_02 = int(input("Masukkan koefisien x pada garis kedua = "))
    e_02 = int(input("Masukkan koefisien y pada garis kedua = "))
    f_02 = int(input("Masukkan konstanta garis kedua = "))
    
    det_02 = a_02 * e_02 - b_02 * d_02
    
    if det_02 == 0:
        print("Kedua garis sejajar atau berhimpitan")
        if (a_02 * f_02 - c_02 * d_02) == 0 and (b_02 * f_02 - c_02 * e_02) == 0:
            print("Kedua garis berhimpitan")
            x_02 = np.linspace(-5, 5, 150)
            y_02 = (c_02 - a_02 * x_02) / b_02
            plt.plot(x_02, y_02, color='blue', label='Garis berhimpitan')
            plt.axvline(x=0, color='red', linestyle='--', linewidth=0.5)
            plt.axhline(y=0, color='red', linestyle='--', linewidth=0.5)
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
        else:
            print("Kedua garis sejajar")
            x_02 = np.linspace(-5, 5, 150)
            y1_02 = (c_02 - a_02 * x_02) / b_02
            y2_02 = (f_02 - d_02 * x_02) / e_02
            plt.plot(x_02, y1_02, color='yellow', label='Garis 1')
            plt.plot(x_02, y2_02, color='green', label='Garis 2')
            plt.axvline(x=0, color='red', linestyle='--', linewidth=0.5)
            plt.axhline(y=0, color='red', linestyle='--', linewidth=0.5)
            plt.legend(loc='best')
            plt.grid(True)
            plt.show()
    else:
        print("Kedua garis berpotongan di satu titik.")
        tp_x_02 = (c_02 * e_02 - b_02 * f_02) / det_02
        tp_y_02 = (a_02 * f_02 - c_02 * d_02) / det_02
        print("Titik potong kedua garis adalah ({},{})".format(tp_x_02, tp_y_02))
        
        x_02 = np.linspace(tp_x_02 - 5, tp_x_02 + 5, 150)
        y1_02 = (c_02 - a_02 * x_02) / b_02
        y2_02 = (f_02 - d_02 * x_02) / e_02
        plt.plot(x_02, y1_02, color='yellow', label='Garis 1')
        plt.plot(x_02, y2_02, color='green', label='Garis 2')
        plt.scatter([tp_x_02], [tp_y_02], color='blue', zorder=5)
        plt.text(tp_x_02, tp_y_02, '({}, {})'.format(tp_x_02, tp_y_02), fontsize=12, ha='right')
        plt.axvline(x=0, color='red', linestyle='--', linewidth=0.5)
        plt.axhline(y=0, color='red', linestyle='--', linewidth=0.5)
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()
else:
    print("Hanya terdapat 3 kasus. Mohon masukkan 1, 2 atau 3 saja.")
```
